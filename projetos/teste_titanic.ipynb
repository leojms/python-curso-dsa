```python
import pandas as pd
import numpy as np

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
dados = pd.read_csv("/content/tested.csv")
```

```python
dados.head()
```

```python
dados = dados.drop(["Name", "Ticket", "Cabin", "Embarked"], axis=1)
```

```python
dados.head()
```

```python
dados=dados.set_index(["PassengerId"])
dados = dados.rename(columns={"Survived":"target"}, inplace=False)
```

```python
dados.head()
```

```python
dados.describe()
```

```python
dados.describe(include=['O'])
```

```python
dados["Sex_F"]= np.where(dados["Sex"]=="female", 1, 0)

dados["PClass_1"]= np.where(dados["Pclass"] == 1, 1, 0)
dados["PClass_2"]= np.where(dados["Pclass"] == 2, 1, 0)
dados["PClass_3"]= np.where(dados["Pclass"] == 3, 1, 0)
```

```python
dados = dados.drop(["Sex", "Pclass"], axis=1)
```

```python
dados.head()
```

```python
dados.isnull().sum()
```

```python
dados.fillna(0, inplace=True)
```

```python
dados.isnull().sum()
```

```python
dados.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(dados.drop(["target"], axis=1),
                                                    dados["target"],
                                                    test_size=0.3,
                                                    random_state=1234)

[{"treino": x_train.shape}, {"teste": x_test.shape}]
```

```python
rndforest = RandomForestClassifier(n_estimators=1000,
                                  criterion='gini',
                                  max_depth=5)
rndforest.fit(x_train, y_train)
```

```python
prob = rndforest.predict_proba(dados.drop("target", axis=1))[:,1]
clas = rndforest.predict(dados.drop("target", axis=1))
```

```python
dados["probabilidade"] = prob
dados["classificação"] = clas
```
