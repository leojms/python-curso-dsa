```python
salario = int(input())

if salario>=0 and salario<=600:
  salario = salario * (1+0.17)
  reaj = salario * 0.17
  perc = 17

elif salario>600 and salario<=900:
  salario = salario * (1+0.13)
  reaj = salario * 0.13
  perc = 13
  
elif salario>900 and salario<=1500:
  salario = salario * (1+0.12)
  reaj = salario * 0.12
  perc = 12
  
elif salario>1500 and salario<=2000:
  salario = salario * (1+0.10)
  reaj = salario * 0.10
  perc = 10
  
else: 
  salario = salario * (1+0.05)
  reaj = salario * 0.05
  perc = 5

salario = str("{:.2f}".format(salario)).replace('.',',')
reaj = str("{:.2f}".format(reaj)).replace('.',',')

print(f"Novo salario: {salario} Reajuste ganho: {reaj} Em percentual: {perc} %")
```
