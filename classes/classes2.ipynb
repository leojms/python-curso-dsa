```python
class circulo:

  pi=3.14

  def __init__(self, raio=5):
    self.raio=raio
  def area(self):
    return (self.raio**2)*circulo.pi
  def set_raio(self, novo_raio):
    self.raio=novo_raio
  def get_raio(self):
    return self.raio
```

```python
circ=circulo()
```

```python
circ.get_raio()
```

```python
circ=circulo(7)
```

```python
circ.get_raio()
```

```python
circ.area()
```

```python
circ.set_raio(7.56)
```

```python
circ.area()
```
