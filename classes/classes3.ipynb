```python
class animal():
  def __init__(self):
    print("Animal criado")
  def identif(self):
    print("Animal")
  def comendo(self):
    print("animal comendo")
```

```python
class cachorro(animal):
  def __init__(self):
    animal.__init__(self)
    print("Objeto cachorro criado")
  def identif(self):
    print("Cachorro")
  def latir(self):
    print("au au")
```

```python
scott=cachorro()
```

```python
scott.identif()
```

```python
scott.latir()
```

```python
scott.comendo()
```
