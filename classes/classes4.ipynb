```python
class livro():
  def __init__(self, nome, autor, paginas):
    self.nome = nome
    self.autor = autor
    self.paginas = paginas
  
  def __str__(self):
    return "Título: %s, Autor: %s, Páginas: %s" \
    %(self.nome, self.autor, self.paginas)

  def __len__(self):
    return self.paginas

  def len(self):
    return print(f"Páginas do livro: {self.paginas}")
```

```python
peq_prin =livro("O Pequeno Príncipe", "Antoine de Saint-Exupéry", 96)
```

```python
print(peq_prin)
```

```python
str(peq_prin)
```

```python
len(peq_prin)
```
