```python
class carro:
  def __init__(self, marca, cor):
    self.marca=marca
    self.cor=cor
  def printar(self):
    print(f"O carro é da marca {self.marca} e da cor {self.cor}")
```

```python
carro1=carro("Volkswagen", "azul")
```

```python
carro1.cor
```

```python
carro1.printar()
```

```python
hasattr(carro1,"marca")
```

```python
setattr(carro1, "marca", "BMW")
```
