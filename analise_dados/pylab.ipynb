```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from pylab import *
```

```python
x=linspace(0,5,10)
y= x**2

fig = plt.figure()
eixos=fig.add_axes([0.1, 0.1, 0.8, 0.8])

eixos.plot(x, y, 'r')
eixos.set_xlabel('x')
eixos.set_ylabel('y')
eixos.set_title('Gráfico')
```

```python
x=linspace(0,5,10)
y= x**2

fig = plt.figure()
eixos=fig.add_axes([0.1, 0.1, 0.8, 0.8])
eixos2=fig.add_axes([0.2, 0.5, 0.4, 0.3])

eixos.plot(x, y, 'r')
eixos.set_xlabel('x')
eixos.set_ylabel('y')
eixos.set_title('Gráfico 1')

eixos2.plot(y, x, 'b')
eixos2.set_xlabel('y')
eixos2.set_ylabel('x')
eixos2.set_title('Gráfico 2')
```
