```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
plt.plot([2, 4, 1], [8, 5, 3])
plt.show()
```

```python
x=[2, 4, 1]
y=[8, 5, 3]
plt.plot(x, y)
plt.xlabel("Eixo x")
plt.ylabel("Eixo y")
plt.title("Teste")
plt.show()
```

```python
x1=[2, 4, 1]
y1=[8, 5, 3]
plt.plot(x1, y1, label="Linha 1")
plt.legend()
plt.show()
```

```python
import numpy as np
```

```python
x=np.arange(1,11,2)
y=np.random.randint(1,10,5)
plt.bar(x, y, label="Barras1", color="r")
plt.legend()
plt.show()
```

```python
x2=np.arange(2,11,2)
y2=np.random.randint(1,10,5)
```

```python
plt.bar(x, y, label="Barras1", color="r")
plt.bar(x2, y2, label="Barras 2", color="y")
plt.legend()
plt.show()
```

```python
idades = np.random.randint(1, 100, 50)
idades = list(idades)
```

```python
ids = [x for x in range(len(idades))]
```

```python
plt.bar(ids, idades)
plt.show
```

```python
bins = list(np.arange(0,101,10))
```

```python
plt.hist(idades, bins, histtype = 'bar', rwidth = 0.8)
plt.show()
```

```python
plt.hist(idades, bins, histtype = 'stepfilled', rwidth = 0.8)
plt.show()
```

```python
x=np.arange(1,9)
y=np.random.randint(1,9,8)
```

```python
plt.scatter(x,y, label='Pontos', color='r',marker='o', s=100)
plt.legend()
plt.show()
```
